```python
# Cell 1
import re
import string
import joblib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
STOPWORDS = set(stopwords.words('english'))

print("Ready. Pandas version:", pd.__version__)

```

```python
# Cell 2
df = pd.read_csv("IMDB Dataset.csv")
df = df[['review','sentiment']].copy()

# sentiment to numeric: positive->1, negative->0
df['sentiment'] = df['sentiment'].map({'positive':1, 'negative':0})

# simple cleaning: remove HTML, lower, remove punctuation, remove extra spaces
def clean_text(text):
    if not isinstance(text, str):
        return ""
    text = re.sub(r"<.*?>", " ", text)                      # remove HTML tags
    text = text.lower()
    text = re.sub(r"http\S+|www\S+|https\S+", " ", text)    # remove urls
    text = re.sub(r"[^a-z0-9\s]", " ", text)                # keep letters + numbers
    text = re.sub(r"\s+", " ", text).strip()
    # optionally remove stopwords (comment out if you want them kept)
    tokens = [w for w in text.split() if w not in STOPWORDS]
    return " ".join(tokens)

# apply cleaning to a small sample for speed-check, then to all
df['clean_review'] = df['review'].apply(clean_text)

# quick check
df.head(3)
```

```python
# Cell 3
X = df['clean_review']
y = df['sentiment']

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, random_state=42, stratify=y)

print("Train size:", len(X_train), "Test size:", len(X_test))
```

```python
# Cell 4
tfidf = TfidfVectorizer(max_features=10000, ngram_range=(1,2))  # 1-grams + 2-grams
X_train_vec = tfidf.fit_transform(X_train)
X_test_vec  = tfidf.transform(X_test)

print("Vectorized shapes:", X_train_vec.shape, X_test_vec.shape)
```

```python
# Cell 5
model = LogisticRegression(max_iter=1000)
model.fit(X_train_vec, y_train)

preds = model.predict(X_test_vec)
print("Accuracy:", accuracy_score(y_test, preds))
print("\nClassification report:\n", classification_report(y_test, preds))
```

```python
# Cell 6
cm = confusion_matrix(y_test, preds)
plt.figure(figsize=(5,4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['neg','pos'], yticklabels=['neg','pos'])
plt.xlabel('Predicted'); plt.ylabel('Actual'); plt.title('Confusion Matrix')
plt.show()
```

```python
# Cell 7
joblib.dump(model, "sentiment_logreg.joblib")
joblib.dump(tfidf, "tfidf_vectorizer.joblib")
print("Saved: sentiment_logreg.joblib and tfidf_vectorizer.joblib")
```

```python
# Cell 8
def predict_review(text):
    text_clean = clean_text(text)
    vec = tfidf.transform([text_clean])
    pred = model.predict(vec)[0]
    proba = model.predict_proba(vec)[0].max()
    return ("positive" if pred==1 else "negative", float(proba))

# Example
example = "This movie was amazing â€” great acting and story!"
print(predict_review(example))
```
